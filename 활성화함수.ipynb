```python
import numpy as np
import matplotlib.pylab as plt
```

```python
# step function
def step_function(x):
    return np.array(x>0, dtype=np.int64)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
```

```python
# sigmoid function
# S자 모양 함수
# h(x) = ( 1 + e ^(-x) ) ^- (1)
```

```python
def sigmoid(x):
    return 1 / (1+np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # y축 범위 지정
plt.show()
```

```python
# ReLU function
# Rectified(정류된) Linear Unit Function (렐루)
# 정류(전기회로 용어) : +/-가 반복되는 교류에서 한쪽 흐름만을 허용
# input > 0 이면 입력을 그대로 출력, input < 0 이면 0을 출력
```

```python
# h(x) = x if x>0
#       0 if x<=0
```

```python
def relu(x):
    return np.maximum(0, x)
# numpy의 maximum은 두 입력 중 큰 값을 선택해 반환하는 함수
```

```python
x = np.arange(-6.0, 6.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-1, 5.5) # y축 범위 지정
plt.show()
```
